```python
from pytrends.request import TrendReq
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ['Buy Bitcoin', 'Sell Bitcoin', 'How to Buy Bitcoin', 'Bitcoin guide']
data = pytrends.get_historical_interest(kw_list, year_start=2018, month_start=1, day_start=1, 
                                 hour_start=0, year_end=2018, month_end=2, day_end=1, 
                                 hour_end=0, cat=0, geo='', gprop='', sleep=10)
```

```python
pytrend = TrendReq()
# Create payload and capture API tokens
pytrend.build_payload(kw_list=['Buy Bitcoin', 'Sell Bitcoin', 'How to Buy Bitcoin', 'Bitcoin guide'], 
                      timeframe='2017-4-1 2017-6-1')

# Interest Over Time
interest_over_time_df = pytrend.interest_over_time()
```
