# Bitcointalk.org forum sentiment analysis
In this notebook we load the data from bitcointalk.org forum (private source not shared on this git), and carry Sentiment Analysis on it.

# 0. Preprocessing

```python
import pandas as pd
import numpy as np
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 15)
```

### 0.3 Bitcointalk forum dataset

```python
forum_data = pd.read_csv('Data/NLP_Cryptocurrency_Dataset.csv', header = 1)
```

```python
forum_data = forum_data.drop('Unnamed: 0', axis = 1)
```

```python
forum_data.datetime = pd.to_datetime(forum_data.datetime)
```

```python
forum_data_prep = forum_data.set_index('datetime').copy()
```

```python
forum_data_prep = forum_data_prep["may-2017":"may-2018"].dropna()
```

```python
forum_data_prep = forum_data_prep.reset_index()
```

```python
forum_data_prep
```

# 1. Sentiment Analysis

https://www.learndatasci.com/tutorials/sentiment-analysis-reddit-headlines-pythons-nltk/

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
```

```python
sia = SIA()
results = []

for line in forum_data_prep.post:  #[:1000]
    pol_score = sia.polarity_scores(line)
    pol_score['post'] = line
    results.append(pol_score)

results[:3]
```

```python
forum_sentiment = pd.DataFrame.from_records(results)
```

```python
forum_sentiment
```

```python
forum_sentiment.to_csv('Data/forum_sentiment.csv')
```

```python
forum_sentiment
```

```python
forum_sentiment.columns = ["compound","neg","neu","pos","comment"]
```

```python
forum_NLP = pd.concat([forum_data_prep, forum_sentiment], axis = 1)
```

```python
forum_NLP = forum_NLP.drop("comment", axis = 1)
```
